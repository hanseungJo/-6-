```python
import pandas as pd
import numpy as np

# github의 자동차 생산 자료를 읽어들인다
path = 'https://github.com/dongupak/DataML/raw/main/csv/'
file = path + 'vehicle_prod.csv'
df = pd.read_csv(file, index_col=0)

# 데이터 프레임에서 국가명을 제외한 숫자의 합을 구하여 'Total'로 명명
sums = df.select_dtypes(np.number).sum()  # 숫자형 데이터의 합계
sums.name = 'Total'  # 'Total' 행으로 설정

# 데이터 프레임에서 국가명을 제외한 숫자의 평균을 구하여 'Average'로 명명
averages = df.select_dtypes(np.number).mean()  # 숫자형 데이터의 평균
averages.name = 'Average'  # 'Average' 행으로 설정

# 데이터 프레임에 'Total'과 'Average' 행을 추가
total_avg_df = pd.concat([df, sums.to_frame().T, averages.to_frame().T])

# 출력
total_avg_df

```
